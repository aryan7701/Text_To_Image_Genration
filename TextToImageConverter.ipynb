```python
from diffusers import DiffusionPipeline
import torch

# Initialize the pipeline
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",  # Model name
    torch_dtype=torch.float16,  # Use 16-bit floats for efficiency
    use_safetensors=True,  # Safe tensor loading
    variant="fp16"  # Use half-precision floating point
)

# Move the model to the GPU
pipe.to("cuda")
```

```python
def generate_image(prompt, pipe):
    # Generate the image
    with torch.no_grad():
        images = pipe(prompt=prompt).images
    return images[0]  # Return the first image (since it's a batch of 1)

# Example prompt
prompt = "Super Nova colorful image"

# Generate the image
image = generate_image(prompt, pipe)

# Save the image
image.save("Black_Hole.png")
```
